DATA CLEANING

```python
##missing values
ds[ds.isnull().any(axis=1)]
```

```python
ds.loc[:122,"Region"]=0
ds.loc[122:,"Region"]=1
```

```python
ds.info()
```

```python
ds[['Region']]=ds[['Region']].astype(int)
```

```python
ds.head()
```

```python
ds.isnull().sum()
```

```python
##removing null values
ds=ds.dropna().reset_index(drop=True)
```

```python
ds.head()
```

```python
ds.isnull().sum()
```

```python
ds=ds.drop(122).reset_index(drop=True)
```

```python
ds.iloc[122]
```

```python
ds.columns
```

```python
## fix spaces in column names
ds.columns.str.strip()
```

```python
ds
```

```python
ds.info()
```

```python
ds[['day','month','year','Temperature',' RH',' Ws']]=ds[['day','month','year','Temperature',' RH',' Ws']].astype(int)
```

```python
ds.info()
```

Changing the other columns to float data datatype

```python
objects=[features for features in ds.columns if ds[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes  ':
        ds[i]=ds[i].astype(float)
```

```python
ds.info()
```

```python
objects
```

```python
##Let ave the cleaned dataset
ds.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

EDA

```python
## drop day,month and year
ds_copy=ds.drop(['day','month','year'],axis=1)
```

```python
ds_copy.head()
```

```python
## categories in classes
ds_copy['Classes  '].value_counts()
```

```python
## Encoding of the categories in classes
ds_copy['Classes  ']=np.where(ds_copy['Classes  '].str.contains('not fire'),0,1)
```

```python
ds_copy.head()
```

```python
ds_copy.tail()
```

```python
## Plot density plot for all features
plt.style.use('ggplot')
ds_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## Percentage for Pie Chart
percentage=ds_copy['Classes'].value_counts(normalize=True)*100
```

```python
# plotting piechart
classlabels=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
